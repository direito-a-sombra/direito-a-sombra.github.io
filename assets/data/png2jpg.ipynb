```python
from os import listdir, path
from PIL import Image as PImage
```

```python
fnames = sorted([f for f in listdir("../images/process0") if f.endswith(("png", "jpg"))])

for f in fnames:
  id = f.split(" ")[0]
  namein = path.join("../images/process0", f)
  nameout = path.join("../images/process", f"{id}.jpg")
  PImage.open(namein).convert("RGB").save(nameout)
```

```python
import csv
import json
```

```python
csvfile = open("../images/process0/subtitles.csv")
reader = csv.reader(csvfile)
data = list(reader)

subtitles = []

for row in data:
  subtitles.append({
    "id": row[0],
    "text": row[1],
  })
```

```python
with open("./process_subtitles.json", "w") as ofp:
  json.dump(subtitles, ofp, ensure_ascii=False)
```
